```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# URL of the Wikipedia page
url = "https://es.wikipedia.org/wiki/Alta_velocidad_ferroviaria_en_Espa%C3%B1a"
```

```python
# Send a GET request to the URL
response = requests.get(url)
```

```python
# Create a BeautifulSoup object
soup = BeautifulSoup(response.text, "html.parser")
```

```python
# Find the table with the desired name
table_name = "Líneas en servicio y en construcción"
table = None
```

```python
# Find the section containing the table
table_section = soup.find("span", id="Líneas_en_servicio_y_en_construcción")
if table_section:
    # Find the next table sibling of the section
    table = table_section.find_next("table")
```

```python
# Extract the data from the table
if table:
    data = []
    for row in table.find_all("tr"):
        cells = [cell.get_text(strip=True) for cell in row.find_all("td")]
        if cells:
            data.append(cells)

    # Print the extracted data
    for row in data:
        print(row)
else:
    print("Table not found.")
```

```python
# Create a DataFrame from the extracted data
columns = ["Línea", "Recorrido (km)", "Estado", "Puesta en servicio"]
df1 = pd.DataFrame(data, columns=columns)

# Save the DataFrame to a CSV file
df1.to_csv("railway_linesfinal.csv", index=False)
```
